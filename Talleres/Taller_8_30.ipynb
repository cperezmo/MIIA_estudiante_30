### Ejercicio 1

### Ejercicio 2
Utiliza `plotly` para realizar un grÃ¡fico que permita verificar si el cambio en el monto de transacciones (cuarto trimestre a primer trimestre) se ajusta a una distribuciÃ³n normal.

```python
# YOUR CODE HERE

x = list(df['Total_Amt_Chng_Q4_Q1'])
y = list(range(len(df['Total_Amt_Chng_Q4_Q1'])))
print(x)
print(y)
```

```python
fig = go.Figure()

fig.add_trace(
    go.Histogram(
        x = x,
        y = y,
        name = "Cambio en el monto de transacciones (cuarto trimestre a primer trimestre)"
    )
)

fig.update_layout(
    title = "Cambio en el monto de transacciones (cuarto trimestre a primer trimestre)",
    xaxis_title = "No. Cliente",
    yaxis_title = "Monto de transacciones",

)

fig.show()
```

### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel mÃ¡ximo de escolaridad alcanzado.
2. NÃºmero de niÃ±os en el hogar.
3. NÃºmero de adolescentes en el hogar.

```python
# YOUR CODE HERE

import ipywidgets as widgets
import pandas as pd

mercadeo = pd.read_csv('/content/marketing_data.csv')
```

```python
def valores_filtro_columna(columna):
  columna = columna.astype(str)
  valores = columna.unique().tolist()
  valores = sorted(valores)
  valores.insert(0,"Todos")
  return valores
```

```python
menu = widgets.Dropdown(options = valores_filtro_columna(mercadeo["Education"]))
```

```python
seleccion = widgets.Output()
```

```python
def accion_cambio_menu(change):
  seleccion.clear_output()

  with seleccion:
    if (change["new"] == "Todos"):
      display(mercadeo)
    else:
      display(mercadeo[mercadeo["Education"] == change["new"]])
```

```python
menu.observe(accion_cambio_menu, names='value')
```

```python
display(menu, seleccion)
```
